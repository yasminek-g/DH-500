As we can observe, the target labels seem to be standardize but not the hero and villain (there are instances of "Trump", "Donald Trump", "trump"), so let's standardize that.

```python
name_map = {
    "Trump": "Donald Trump",
    "Donald": "Donald Trump",
    "Biden": "Joe Biden",
    "Harris": "Kamala Harris",
    "Kamala": "Kamala Harris",
    "Musk": "Elon Musk",
    "Elon": "Elon Musk",
    "Obama": "Barack Obama",
    "Ai": "AI",
}

def standardize_entity(entity):
    if not isinstance(entity, str):
        return entity  # Handle unexpected non-string values gracefully

    # Split multi-name entries
    entity = entity.title()
    entity = entity.strip()
    names = entity.split(" And ")
    standardized_names = [name_map.get(name, name) for name in names]

    return " and ".join(sorted(set(standardized_names))).replace("President ", "").replace("Vp ", "")
```

```python
df['hero'] = df['hero'].apply(standardize_entity)
df['hero'].value_counts().head(20)
```

```python
df['villain'] = df['villain'].apply(standardize_entity)
df['villain'].value_counts().head(20)
```
